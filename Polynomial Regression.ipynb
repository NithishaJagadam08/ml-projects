# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
data=pd.read_csv('/content/Corn_Yield.csv')
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=2)
x_poly=pf.fit_transform(x)
poly_reg=LinearRegression()
poly_reg.fit(x_poly,y)
```

## Visualising the Linear Regression results

```
plt.scatter(x,y,color='blue')
plt.plot(x,model.predict(x),color='yellow')
plt.title('Linear Regression for crop Yield')
plt.xlabel('fertilizer usd')
plt.ylabel('corn Yield')
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(x,y,color='red')
plt.plot(x,poly_reg.predict(x_poly),color='black')
plt.title('Polynomial Regression for crop Yield')
plt.xlabel('fertilizer usd')
plt.ylabel('corn Yield')
plt.show()
```

## Predicting a new result with Linear Regression

```
model.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```
poly_reg.predict(pf.fit_transform([[6.5]]))
```
